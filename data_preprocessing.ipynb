```
listfile = ['/content/iran_092020_tweets_csv_hashed.csv',
            '/content/iran_201906_1_tweets_csv_hashed.csv',
            '/content/iran_201906_2_tweets_csv_hashed.csv',
            '/content/iran_201906_3_tweets_csv_hashed.csv']
dfx = pd.DataFrame()

def clean_text(df2):
  
  df2['tweet_text']=df2['tweet_text'].map(lambda x: x.lstrip('RT'))
  df2['tweet_text'] = df2['tweet_text'].map(lambda x: re.sub("@[A-Za-z0-9_:\+\=]+","",x))
  df2['tweet_text'] = df2['tweet_text'].map(lambda x: re.sub('http\S+|www.\S+', '', x))
  df2['tweet_text'] = df2['tweet_text'].map(lambda x: re.sub('[()]+','', x))
  df2.tweet_text = df2.tweet_text.map(lambda x: re.sub("([^\x00-\x7F])+","",x))
  df2['tweet_text']=df2['tweet_text'].map(lambda x: x.strip())
  for index in df2.index:
    text = df2['tweet_text'][index]
    if all(i in string.punctuation for i in text):
      df2['tweet_text'][index]=''
  
  df2 = df2[df2['tweet_text'].str.split().str.len().gt(2)]
  nan_value = float("NaN")
  df2.replace("", nan_value, inplace=True)
  df2.dropna(subset = ["tweet_text"], inplace=True)
  df2.to_csv('result.csv', mode='a', header=False)
 
for f in listfile:
    print(f)
    for df in pd.read_csv(f, sep=',',usecols= ['tweet_text'], iterator=True, chunksize=1000):
      clean_text(df)
     
```
